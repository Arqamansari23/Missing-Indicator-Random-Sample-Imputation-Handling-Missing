# Missing Indicator


We can create Seperate Column in which defines the presence of null values  

So, Our model can identify the Null values and work Efficiently  

# Importing Library

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split

from sklearn.impute import MissingIndicator,SimpleImputer
```

# Importing Data set

```python
df = pd.read_csv('train.csv',usecols=['Age','Fare','Survived'])
df.head()
```

# Train_Test_split 

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
X_train.head()
```

# First Check Accuracy With Simple Imputer 

```python
si = SimpleImputer()
X_train_trf = si.fit_transform(X_train)
X_test_trf = si.transform(X_test)
```

```python
X_train_trf          
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression()

clf.fit(X_train_trf,y_train)

y_pred = clf.predict(X_test_trf)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

From Simple Imputer Accuracy Is 61%

# Check Accuracy With Missing Imputer 

```python
mi = MissingIndicator()

mi.fit(X_train)  # fitting Missing Indicator on Training Data
```

```python
X_train_missing = mi.transform(X_train)  # transform on both Train and test data 
X_test_missing = mi.transform(X_test)
```

```python
X_train_missing  # where Data is missing it gives True
```

Joining X_train_missing colum Into data frame 

```python
X_train['Age_NA']= X_train_missing
X_test['Age_NA']= X_test_missing
```

```python
X_train.sample(5)  # Data frame joined 
```

Now Applying Simple Imputer again and check accuracy

```python
si = SimpleImputer()
X_train_trf2 = si.fit_transform(X_train)
X_test_trf2 = si.transform(X_test)

clf = LogisticRegression()

clf.fit(X_train_trf2,y_train)

y_pred = clf.predict(X_test_trf2)

from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

We can Clearly See that Accuracy is Increased By 2 %
